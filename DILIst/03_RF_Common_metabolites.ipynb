```python
df_unique_compounds= df[df["Input_Standardised_SMILES"]==df["Output"]].drop_duplicates(subset=["InchiKey"], keep="first").reset_index(drop=True)
df_unique_compounds
```

```python
# Step 1: Group by compound and collect unique metabolites for each

df_toxic_compounds= df[df["Input_Standardised_SMILES"]!=df["Output"]]
df_toxic_compounds = df_toxic_compounds[df_toxic_compounds.Input_Standardised_SMILES.isin(df_unique_compounds.Input_Standardised_SMILES.to_list())]
df_toxic_compounds = df_toxic_compounds[df_toxic_compounds["Label"]==1]
#Drop duplicates, conservative approach to keep evidence of toxicity
df_toxic_compounds = df_toxic_compounds
print(len(df_toxic_compounds))

compound_to_metabolites = df_toxic_compounds.groupby('Input_SMILES')['InchiKey'].apply(set).to_dict()
```

```python
# Step 2: Create a reverse mapping from metabolites to compounds
metabolite_to_compounds = {}
for compound, metabolites in compound_to_metabolites.items():
    for metabolite in metabolites:
        if metabolite not in metabolite_to_compounds:
            metabolite_to_compounds[metabolite] = set()
        metabolite_to_compounds[metabolite].add(compound)
        
```

```python
# Step 3: For each compound, find common metabolites shared with other compounds
compound_common_metabolites_count = {}
for compound, metabolites in compound_to_metabolites.items():
    common_metabolites = set()
    for metabolite in metabolites:
        # Accumulate compounds that share the same metabolite, excluding the current one
        common_metabolites.update(metabolite_to_compounds[metabolite] - {compound})
    compound_common_metabolites_count[compound] = len(common_metabolites)
```

```python

# Step 4: Report the findings
for compound, count in compound_common_metabolites_count.items():
    if (count>0):
        print(f"Compound {compound} shares common metabolites with {count} other compounds.")
```

```python
metabolite_toxic_compound_counts = {metabolite: len(compounds) for metabolite, compounds in metabolite_to_compounds.items()}

for metabolite, count in metabolite_toxic_compound_counts.items():
    if count > 1:  # If the metabolite is shared by any toxic compounds
        print(f"Metabolite {metabolite} is present in {count} toxic compounds.")    
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Function to convert SMILES string to Morgan fingerprint
def smiles_to_fingerprint(smiles, n_bits=2048):
    mol = Chem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=n_bits)
    return list(fp)

# Apply the function to the 'smiles' column and create a new column for the fingerprint
df['morgan_fingerprint'] = df['protonated_Output'].apply(smiles_to_fingerprint)
```

```python
df_DILI_unique_compounds= df[df["Input_Standardised_SMILES"]==df["Output"]].drop_duplicates(subset=["InchiKey"], keep="first").reset_index(drop=True)
df_DILI_unique_compounds
```

```python
df_labelled = df[df.Input_Standardised_SMILES.isin(df_DILI_unique_compounds.Input_Standardised_SMILES.to_list())]
df_labelled
```

```python
X = df_DILI_unique_compounds.Input_SMILES.to_list()
y = df_DILI_unique_compounds.Label.to_list()

from sklearn.model_selection import train_test_split

X_train_smiles, X_test_smiles, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=24, stratify=y)

print(len(X_train_smiles))
print(len(X_test_smiles))
```

```python
df_train = df[df.Input_SMILES.isin(X_train_smiles)]
df_train = df_train[df_train["Input_Standardised_SMILES"]!=df_train["Output"]].reset_index(drop=True)
df_test = df[df.Input_SMILES.isin(X_test_smiles)]
df_test = df_test[df_test["Input_Standardised_SMILES"]!=df_test["Output"]].reset_index(drop=True)


print(len(df_train))
print(len(df_test))
```

```python
# Step 1: Identify common metabolites
common_metabolites = set(df_train['InchiKey']).intersection(set(df_test['InchiKey']))
print(len(common_metabolites))
```

```python
import numpy as np
```

```python
from tqdm import tqdm

# Step 2: Check if the labels agree for these common metabolites
disagreements = []  # Store instances where labels disagree
# Initialize storage for agreement cases
agreements = []

# Track the details of common metabolites, including their associated compounds and labels
for metabolite in tqdm(common_metabolites):
    train_entries = df_train[df_train['InchiKey'] == metabolite]
    test_entries = df_test[df_test['InchiKey'] == metabolite]
    
    train_labels = train_entries['Label'].unique()
    test_labels = test_entries['Label'].unique()
    
    # Check for label disagreement
    if not np.array_equal(train_labels, test_labels):
        disagreements.append((
            metabolite, 
            train_labels.tolist(), 
            test_labels.tolist(),
            train_entries[['Input_SMILES', 'Label']].values.tolist(),
            test_entries[['Input_SMILES', 'Label']].values.tolist()
        ))
    else:  # Labels agree
        agreements.append((
            metabolite, 
            train_labels.tolist(), 
            test_labels.tolist(),
            train_entries[['Input_SMILES', 'Label']].values.tolist(),
            test_entries[['Input_SMILES', 'Label']].values.tolist()
        ))

# Report findings
if disagreements:
    print(f"Found {len(disagreements)} common metabolites with label disagreements.")
    for metabolite, train_labels, test_labels, train_compounds, test_compounds in disagreements:
        print(f"Metabolite {metabolite} disagrees. Train labels: {train_labels}, Test labels: {test_labels}.")
        print(f"Train compounds and labels: {train_compounds}")
        print(f"Test compounds and labels: {test_compounds}")
else:
    print("No disagreements found among common metabolites.")

if agreements:
    print(f"\nFound {len(agreements)} common metabolites with label agreements.")
    for metabolite, train_labels, test_labels, train_compounds, test_compounds in agreements:
        print(f"Metabolite {metabolite} agrees. Labels: {train_labels}.")
        print(f"Train compounds and labels: {train_compounds}")
        print(f"Test compounds and labels: {test_compounds}")
else:
    print("No agreements found among common metabolites.")
```

```python
df_train = df[df.Input_SMILES.isin(X_train_smiles)]
df_test = df[df.Input_SMILES.isin(X_test_smiles)]

'''
Do not consider mertabolites

df_train = df_train[df_train["Input_Standardised_SMILES"]==df_train["Output"]].reset_index(drop=True)
df_test = df_test[df_test["Input_Standardised_SMILES"]==df_test["Output"]].reset_index(drop=True)
'''

X_train = np.array(df_train.morgan_fingerprint.tolist())
y_train = df_train.Label.to_list()

X_test = np.array(df_test.morgan_fingerprint.tolist())
y_test = df_test.Label.to_list()

print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import balanced_accuracy_score, roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_distributions = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'n_jobs': [10],
    'random_state': [42]
}

clf = RandomForestClassifier(random_state=42)

searcher = RandomizedSearchCV(clf, param_distributions, n_iter=10, scoring='balanced_accuracy', 
                                      cv=5, n_jobs =-1, random_state=42)
searcher.fit(X_train, y_train)

# Update clf to the best estimator
clf = searcher.best_estimator_

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class

```

```python
df_test["preds"] = y_proba
```

```python
new_df_voting = df_test[['Input_SMILES', 'Label','preds']].groupby("Input_SMILES").median().reset_index()
new_df_voting
```

```python
y_proba = new_df_voting.preds
y_pred = new_df_voting.round().preds
y_test = new_df_voting.Label

balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
auc_score = roc_auc_score(y_test, y_proba)  # Ensure y_test is appropriately encoded for binary classification

print(f'The balanced accuracy score is {balanced_accuracy}.')
print(f'The AUC score is {auc_score}.')
```
