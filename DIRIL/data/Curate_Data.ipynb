```python
import pandas as pd
```

```python
df = pd.read_csv("DIRIL.csv")
df
```

```python
# Binarize the 'Type' column
df['DIRIL_label'] = df['DIRIL_Rank'].apply(lambda x: 1 if x == 'Nephrotoxic' else 0)
df
```

```python
df.DIRIL_label.value_counts()
```

```python
# Attempt conversion where possible, otherwise copy the values
def convert_or_copy(row):
    # Attempt to convert to float for numerical operations
    try:
        quantity = float(row['2023 DDD'])
        # Convert from mg to g if applicable
        if row['2023 Unit'] == 'mg':
            return quantity / 1000, 'g'
        else:
            return quantity, 'g'  # No conversion needed but ensure unit is standardized to 'g'
    except ValueError:
        # Non-convertible values are copied directly
        return row['2023 DDD'], row['2023 Unit']

# Apply the conversion or copy logic
df[['2023 DDD', '2023 Unit']] = df.apply(lambda row: convert_or_copy(row), axis=1, result_type='expand')

df
```

```python
df.to_csv("DIRIL_curated.csv", index=False)
```
