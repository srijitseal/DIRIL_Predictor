```python
import pandas as pd
```

```python
df = pd.read_csv("nephrotox_all.csv")
df
```

```python
# Revising the approach to avoid TypeError with NA values
# Using a function that explicitly handles NA values to avoid ambiguity in boolean expressions

def transform_label_gong(x):
    if pd.isna(x):
        return None  # Explicitly handling NA values
    elif x == 'Nephrotoxic':
        return 1
    elif x == 'Non-nephrotoxic':
        return 0
    else:
        return None  # Handling unexpected values

def transform_label_shi(x):
    if pd.isna(x):
        return None  # Explicitly handling NA values
    elif x == 'nephrotoxic':
        return 1
    elif x == 'non-nephrotoxic':
        return 0
    else:
        return None  # Handling unexpected values

# Applying the transformation functions
df['Label_Gong'] = df['Label_Gong'].apply(transform_label_gong)
df['Label_Shi'] = df['Label_Shi'].apply(transform_label_shi)

df
```

```python
filtered_df_direct = df[((df['Label_Gong'].isna() | df['Label_Shi'].isna()) | (df['Label_Gong'] == df['Label_Shi']))].reset_index(drop=True)
filtered_df_direct
```

```python
df['Label_Shi'].value_counts()
```

```python
# Binarize the 'Type' column
df['Label_Gong'] = df['Label_Gong'].apply(lambda x: 1 if x == 'Nephrotoxic' else 0)
df['Label_Shi'] = df['Label_Gong'].apply(lambda x: 1 if x == 'nephrotoxic' else 0)
df
```

```python
# Binarize the 'Type' column
df['DIRIL_label'] = df['DIRIL_Rank'].apply(lambda x: 1 if x == 'Nephrotoxic' else 0)
df
```

```python
df.DIRIL_label.value_counts()
```

```python
# Attempt conversion where possible, otherwise copy the values
def convert_or_copy(row):
    # Attempt to convert to float for numerical operations
    try:
        quantity = float(row['2023 DDD'])
        # Convert from mg to g if applicable
        if row['2023 Unit'] == 'mg':
            return quantity / 1000, 'g'
        else:
            return quantity, 'g'  # No conversion needed but ensure unit is standardized to 'g'
    except ValueError:
        # Non-convertible values are copied directly
        return row['2023 DDD'], row['2023 Unit']

# Apply the conversion or copy logic
df[['2023 DDD', '2023 Unit']] = df.apply(lambda row: convert_or_copy(row), axis=1, result_type='expand')

df
```

```python
df.to_csv("DIRIL_curated.csv", index=False)
```
