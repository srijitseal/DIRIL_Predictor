```python
# Revising the approach to avoid TypeError with NA values
# Using a function that explicitly handles NA values to avoid ambiguity in boolean expressions

def transform_label_gong(x):
    if pd.isna(x):
        return None  # Explicitly handling NA values
    elif x == 'Nephrotoxic':
        return 1
    elif x == 'Non-Nephrotoxic':
        return 0
    else:
        return None  # Handling unexpected values

def transform_label_shi(x):
    if pd.isna(x):
        return None  # Explicitly handling NA values
    elif x == 'nephrotoxic':
        return 1
    elif x == 'non-nephrotoxic':
        return 0
    else:
        return None  # Handling unexpected values

# Applying the transformation functions
df['Label_Gong'] = df['Label_Gong'].apply(transform_label_gong)
df['Label_Shi'] = df['Label_Shi'].apply(transform_label_shi)

df
```

```python
filtered_df_direct = df[((df['Label_Gong'].isna() | df['Label_Shi'].isna()) | (df['Label_Gong'] == df['Label_Shi']))].reset_index(drop=True)
filtered_df_direct
```

```python
import numpy as np

# Use numpy.where to create new columns based on the availability of values
filtered_df_direct['smiles'] = np.where(filtered_df_direct['SMILES_Shi'].isnull(), filtered_df_direct['SMILES_Gong'], filtered_df_direct['SMILES_Shi'])
filtered_df_direct['label'] = np.where(filtered_df_direct['Label_Shi'].isnull(), filtered_df_direct['Label_Gong'], filtered_df_direct['Label_Shi'])
filtered_df_direct
```

```python
filtered_df_direct["InChI14"] = filtered_df_direct["InCHI_Term"].str[:14]
filtered_df_direct
```

```python
filtered_df_direct = filtered_df_direct.drop_duplicates(["InChI14", "label"], keep=False)
filtered_df_direct = filtered_df_direct.drop_duplicates(["InChI14"], keep="first").reset_index(drop=True)
filtered_df_direct
```

```python
dirilrank = pd.read_csv("../DIRIL_processed.csv")
dirilrank = dirilrank[dirilrank["Input_Standardised_SMILES"] == dirilrank["Output"]]
dirilrank
```

```python
external_test = filtered_df_direct[~filtered_df_direct.InChI14.isin(dirilrank.InchiKey.to_list())]
external_test = external_test[["InChI14", "smiles", "label"]].reset_index(drop=True)
external_test
```
